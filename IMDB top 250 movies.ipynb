```python
#importing the necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
#url
url="https://www.imdb.com/chart/top/"
```

```python
r=requests.get(url,"lxml")
soup=BeautifulSoup(r.content)
```

```python
#getting the titles,year of release and ratings and storing them in 2 arrays
title_and_year=[i.text.strip() for i in soup.findAll("td",{"class":"titleColumn"})]
ratings=[i.text.strip() for i in soup.findAll("td",{"class":"ratingColumn imdbRating"})]
```

```python
title_and_year
```

This format is very clumsy, so we need to extract the movie names and the year of release from it

```python
#extracting the movie names and years of release
movie_name=[]
year=[]
for i in title_and_year:
    r=re.findall(r"\d+.\s+(.+)\n",i)
    movie_name.append(r[0].strip())
    r=re.findall(r"\d{4}",i)
    year.append(r[-1])
```

```python
df=pd.DataFrame(list(zip(movie_name,year,ratings)))
df.columns=["Movie_title","Year of release","Rating"]
df
```

```python
df.dtypes
```

```python
#changing the datatypes so that it can be furthur used for analysis
df["Year of release"]=df["Year of release"].astype("int")
df["Rating"]=df["Rating"].astype("float")
df.dtypes
```

```python
df.to_csv("Top250IMDBMovies.csv",index=False)
```
